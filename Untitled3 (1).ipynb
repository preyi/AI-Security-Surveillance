```python
import cv2
from deepface import DeepFace

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Try different camera indices
for camera_index in [0, 1, 2]:  # Check indices 0, 1, and 2
    cap = cv2.VideoCapture(camera_index)
    if cap.isOpened():
        break  # Exit loop if camera is opened successfully

if not cap.isOpened():
    raise IOError("Cannot open webcam. Please check your camera connection and permissions.")

while True:
    ret, frame = cap.read()
    # analyzing
    result = DeepFace.analyze(frame, actions=['emotion'])
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Use frame instead of all for processing
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # for multiple faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        font = cv2.FONT_HERSHEY_SIMPLEX

        # Accessing dominant emotion from 'result'
        if result and len(result) > 0:
            dominant_emotion = result[0]['dominant_emotion']
            cv2.putText(frame, dominant_emotion, (x, y - 10), font, 1, (0, 0, 255), 2, cv2.LINE_4)

    cv2.imshow("Original Video", frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
